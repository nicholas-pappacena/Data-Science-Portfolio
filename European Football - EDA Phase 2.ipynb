# England - Premier Leage Data (2017 - 2021)

```python
# 2020/2021 Season Data

# Premier League Table
prem_table = pd.read_html('https://fbref.com/en/comps/9/10728/2020-2021-Premier-League-Stats')[0]

cols_to_drop = ['Rk','xG', 'GF', 'GA', 'xGA', 'xGD', 'xGD/90', 'Attendance', 'Top Team Scorer', 'Goalkeeper', 'Notes']
prem_table = prem_table.drop(columns=cols_to_drop)
prem_table.rename(columns={"MP": "Matches", "W": "Wins", "D":"Ties", "L": "Loss", "GD": "Goal Differential", "Pts": "Points"}, inplace = True)
# prem_table.info()

# Add Premier League Goalkeeping, Shooting, Passing, Defending tables

prem_goalie = pd.read_html('https://fbref.com/en/comps/9/10728/2020-2021-Premier-League-Stats')[4]
prem_score = pd.read_html('https://fbref.com/en/comps/9/10728/2020-2021-Premier-League-Stats')[8]
prem_pass = pd.read_html('https://fbref.com/en/comps/9/10728/2020-2021-Premier-League-Stats')[10]
prem_defend = pd.read_html('https://fbref.com/en/comps/9/10728/2020-2021-Premier-League-Stats')[16]

# Clean Goalkeeping Columns

prem_goalie = prem_goalie.drop(columns=[('Unnamed: 1_level_0'), ('Playing Time'), ('Penalty Kicks')])
prem_goalie.columns = prem_goalie.columns.droplevel()
prem_goalie = prem_goalie.drop(columns=['GA90', 'W', 'D', 'L', 'CS%'])
prem_goalie.rename(columns={"GA": "Goals Allowed", 'SoTA': 'Shots on Target Allowed', 'CS': 'Clean Sheets'}, inplace = True)
# prem_goalie.info()

# Clean Shooting Columns

prem_score = prem_score.drop(columns=[('Expected')])
prem_score.columns = prem_score.columns.droplevel()
prem_score = prem_score.drop(columns=['# Pl', '90s', 'Dist', 'Sh/90', 'SoT/90'])
prem_score.rename(columns={'Gls':'Goals Scored', 'Sh':'Shots', 'SoT':'Shots on Target', 'SoT%':'Shots on Target %', 'G/Sh':'Goals Per Shot', 'G/SoT':'Goals Per Shot on Target', 'FK':'Free Kicks', 'PK':'Penalty Kicks Made', 'PKatt':'Penalty Kick Attempted'}, inplace = True)
# prem_score.info()

# Clean Passing Columns

prem_pass.columns = prem_pass.columns.droplevel()
prem_pass = prem_pass.drop(columns=['# Pl', '90s', 'PrgDist', 'xA', 'A-xA', 'Prog'])
prem_pass.columns.values[1] = "Passes Comp_Total"
prem_pass.columns.values[2] = "Passes Att_Total"
prem_pass.columns.values[3] = "Pass Comp %_Total"
prem_pass.columns.values[4] = "Passes Dist_Total"
prem_pass.columns.values[5] = "Passes Comp_Short"
prem_pass.columns.values[6] = "Passes Att_Short"
prem_pass.columns.values[7] = "Passes Comp %_Short"
prem_pass.columns.values[8] = "Passes Comp_Med"
prem_pass.columns.values[9] = "Passes Att_Med"
prem_pass.columns.values[10] = "Passes Comp %_Med"
prem_pass.columns.values[11] = "Passes Comp_Long"
prem_pass.columns.values[12] = "Passes Att_Long"
prem_pass.columns.values[13] = "Passes Comp %_Long"
prem_pass.columns.values[14] = "Assists"
prem_pass.columns.values[15] = "Key Passes" # passes that lead to shots
prem_pass.columns.values[16] = "Final Third"
prem_pass.columns.values[17] = "Comp Pass in Penalty Area"
prem_pass.columns.values[18] = "Comp Crosses"
# prem_pass.info()

# Clean Defending Columns

prem_defend = prem_defend.drop(columns=[('Pressures'),('Vs Dribbles')])
prem_defend.columns = prem_defend.columns.droplevel()
prem_defend = prem_defend.drop(columns=['# Pl', '90s', 'ShSv', 'Clr', 'Tkl+Int'])
prem_defend.rename(columns={"Tkl": "Tackles", 'TklW': 'Tackles Won', 'Def 3rd': 'Tackles in Def 3rd', 'Mid 3rd':'Tackles in Mid 3rd', 'Att 3rd':'Tackles in Att 3rd', 'Sh':'Shots Blocked', 'Pass':'Passes Blocked', 'Int':'Interceptions', 'Err':'Defensive Errors'}, inplace = True)
prem_defend['Tackle Win %'] = round(prem_defend['Tackles Won'] / prem_defend['Tackles'], 2)
# prem_defend.info()

# Merge Data
england_20_21 = prem_table.merge(prem_goalie, on='Squad')
england_20_21 = england_20_21.merge(prem_score, on='Squad')
england_20_21 = england_20_21.merge(prem_pass, on='Squad')
england_20_21 = england_20_21.merge(prem_defend, on='Squad')
england_20_21['Season'] = 'Premier League 20-21'
# england_20_21.head()
```

# Add "Points Per Game" Field

- The Premier League (England), La Liga (Spain), Serie A (Italy), and Ligue 1 (France) all contain 20 teams per league, meaning that each team pays 38 games a season.  
- The Bundesliga (Germany) contains 18 teams, meaning that each team only plays 34 games a season.
- To account for the missed opportunity that Bundesliga teams have in terms of total points earned (as these teams play 4 less games per season), a better comparison across the leagues would be to compare the "Points Per Game" each team across all leagues haas earned as opposed to just the total number of points each team has earned.
- We see that over the past 4 seasons, of the top 10 overall points per game seasons, Manchester City (England), Liverpool (England), Paris Saint Germain (France), and Bayern Munich (Germany) each appear on this list twice.

```python
df['Points Per Game'] = round(df['Points'] / df['Matches'], 2)

df.shape
df.info()
df.head()
```

```python
sec_col_name = "League"
third_col_name = "Season"
fourth_col_name = "Points Per Game"

sec_col = df.pop(sec_col_name)
third_col = df.pop(third_col_name)
fourth_col = df.pop(fourth_col_name)


df.insert(1,sec_col_name, sec_col)
df.insert(2,third_col_name, third_col)
df.insert(3,fourth_col_name, fourth_col)

df.head()
```

```python
df['Comp Pass Final 3rd / Game'] = df['Passes Comp Final 3rd'] / df['Matches']
df['Pass Dist_Total / Game'] = df['Passes Dist_Total'] / df['Matches']
df['Key Passes / Game'] = df['Key Passes'] / df['Matches']
df['Saves / Game'] = df['Saves'] / df['Matches']
df['Tackles in Def 3rd / Game'] = df['Tackles in Def 3rd'] / df['Matches']
df['Shots Blocked / Game'] = df['Shots Blocked'] / df['Matches']

best_teams = df.groupby(['Squad', 'League'])[['Points Per Game','Comp Pass Final 3rd / Game','Pass Dist_Total / Game','Key Passes / Game','Saves / Game','Tackles in Def 3rd / Game','Shots Blocked / Game']].mean().sort_values(by='Points Per Game',ascending = False)[:10]
best_teams = round(best_teams, 2)
best_teams.style.format('{:,}')

round(best_teams['Comp Pass Final 3rd / Game'].mean(),2)
round(best_teams['Pass Dist_Total / Game'].mean(),2)
round(best_teams['Key Passes / Game'].mean(),2)
round(best_teams['Saves / Game'].mean(),2)
round(best_teams['Tackles in Def 3rd / Game'].mean(),2)
round(best_teams['Shots Blocked / Game'].mean(),2)


worst_teams = df.groupby(['Squad', 'League'])[['Points Per Game','Comp Pass Final 3rd / Game','Pass Dist_Total / Game','Key Passes / Game','Saves / Game','Tackles in Def 3rd / Game','Shots Blocked / Game']].mean().sort_values(by='Points Per Game',ascending = True)[:10]
worst_teams = round(worst_teams, 2)
worst_teams.style.format('{:,}')

round(worst_teams['Comp Pass Final 3rd / Game'].mean(),2)
round(worst_teams['Pass Dist_Total / Game'].mean(),2)
round(worst_teams['Key Passes / Game'].mean(),2)
round(worst_teams['Saves / Game'].mean(),2)
round(worst_teams['Tackles in Def 3rd / Game'].mean(),2)
round(worst_teams['Shots Blocked / Game'].mean(),2)
```
